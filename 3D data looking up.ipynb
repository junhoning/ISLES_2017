```python
import cv2
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
from glob import glob
import os
import nibabel as nib
%matplotlib inline
```

```python
data_path = '../../data/ISLES_2017/Training/'
```

```python
image_list = glob(data_path+'*/*ADC*/*.nii')
```

```python
def read_img(image):
    return nib.load(image).get_data().astype(int)
```

```python
for img in image_list:
    image = read_img(img)
    print(image.shape)
```

```python
shape_list = []
for img in image_list:
    image = read_img(img)
    if image.shape not in shape_list:
        shape_list.append(image.shape)
print(shape_list)
```

```python
image = read_img(image_list[0])
print(image.shape)
plt.imshow(image[80])
```

```python
image = read_img(image_list[0])
print(image.shape)
for i in range(9):
    plt.imshow(image.transpose(2,0,1)[i])
    plt.show()
```

```python
from skimage import measure
```

```python
image = read_img(image_list[0])

# Remove background
cent_img_n = image.shape[2] // 2
cnt_image = image.transpose(2,0,1)[cent_img_n]

classes = measure.label(cnt_image)
labels = np.where(classes==0, 0, 1)
new = measure.regionprops(labels)
min_row, min_col, max_row, max_col = new[0].bbox
labled_img = image[min_row:max_row, min_col:max_col]

for i in range(9):
    plt.imshow(image.transpose(2,0,1)[i][min_row:max_row, min_col:max_col])
    plt.show()
```
